## Analyzing all football league table from 2012/13 to 2023/24 season
- Purpose of this code is to find the position of every league & teams per season.
- This data also helps in identification of teams that qualify for general international competitions.
- Provides data about winning teams in every league and in every season for 10 seasons.

```python
competitions = pd.read_csv('competitions.csv')
```

```python
competitions = competitions.drop(['url'], axis=1)
```

```python
competitions.info()
```

```python
competitions[competitions['country_name'].isna()]
competitions['country_name'] = competitions['country_name'].fillna('General')
competitions['domestic_league_code'] = competitions['domestic_league_code'].fillna('GN1')
```

```python
general_competitions = competitions[competitions['country_name'] == 'General']
```

```python
general_competitions['sub_type'] = general_competitions['sub_type'].str.replace('_',' ')
```

```python
general_competitions['sub_type'] = general_competitions['sub_type'].str.title()
```

```python
general_competitions.rename(columns={'name':'sub_type','sub_type':'Name'}, inplace=True)
```

```python
general_competitions['country_id'] = general_competitions['country_id'].replace(-1,0)
general_competitions['confederation'] = general_competitions['confederation'].replace('europa','Europe')
```

```python
general_competitions.drop('sub_type', axis=1, inplace=True)
```

```python
general_competitions.set_index('competition_id', inplace=True)
```

```python
major_leagues = competitions[competitions['is_major_national_league']]
```

```python
major_leagues.drop(['competition_code'], axis=1, inplace=True)
```

```python
major_leagues['name'] = major_leagues['name'].str.title()
```

```python
major_leagues['confederation'] = major_leagues['confederation'].str.replace('europa','Europe')
```

```python
major_leagues.set_index('competition_id', inplace=True)
```

```python
major_leagues.sort_values(by=['country_name'], inplace=True)
```

```python
games = pd.read_csv('games.csv')
```

```python
games_shrink = major_leagues.merge(games, how='left', right_on='competition_id', left_on='domestic_league_code')
```

```python
games_shrink.info()
```

```python
game_details = games_shrink.loc[:,['country_name','name','season','date','round','home_club_name',
                    'away_club_name','aggregate','home_club_goals','away_club_goals','home_club_id','away_club_id']]
```

```python
game_details.sort_values(by=['date','country_name'], inplace=True)
```

```python
def result(row):
    home_goals = row['home_club_goals']
    away_goals = row['away_club_goals']

    if home_goals > away_goals:
        return 'win'
    elif home_goals == away_goals:
        return 'draw'
    else:
        return 'loss'


game_result = game_details.apply(result, axis=1).to_frame().rename(columns={0:'game_result'})
```

```python
game_details = game_details.merge(game_result, how='inner', left_index=True, right_index=True)
```

```python
def h_points(row):
    result = row['game_result']

    if result == 'win':
        return 3
    elif result == 'draw':
        return 1
    else:
        return 0



def a_points(row):
    result = row['game_result']

    if result == 'win':
        return 0
    elif result == 'draw':
        return 1
    else:
        return 3
        
h_game_points = game_details.apply(h_points, axis=1).to_frame().rename(columns={0:'home_game_points'})
a_game_points = game_details.apply(a_points, axis=1).to_frame().rename(columns={0:'away_game_points'})
```

```python
game_details = game_details.merge(h_game_points, how='inner', left_index=True, right_index=True)
game_details = game_details.merge(a_game_points, how='inner', left_index=True, right_index=True)
```

```python
game_details['goal_difference'] = game_details['home_club_goals'] - game_details['away_club_goals']
game_details.head()
```

```python
game_details[(game_details['season'] == 2023) & 
((game_details['home_club_id'] == 31) | (game_details['away_club_id'] == 31))].head()
```

```python
tornament = game_details.loc[:,['country_name','name','season',
                                'home_club_name','away_club_name',
                                'home_club_id','away_club_id',
                                'home_game_points','away_game_points','goal_difference']]
tornament.sort_values(by=['country_name','season','home_club_name','away_club_name'], inplace=True)
```

```python
home_tornament = tornament.loc[:,['season','home_club_name','home_club_id','home_game_points',
                                  'goal_difference']]
# home_tornament.set_index(['season','home_club_id'], inplace=True)
```

```python
home_tornament_summary = home_tornament.pivot_table(values=['home_game_points','goal_difference'], 
                                                    index=['season','home_club_id'], aggfunc='sum')
```

```python
away_tornament = tornament.loc[:,['season','away_club_name','away_club_id','away_game_points',
                                  'goal_difference']]
away_tornament.head(2)
```

```python
away_tornament_summary = away_tornament.pivot_table(values=['away_game_points','goal_difference'], 
                                                    index=['season','away_club_id'], aggfunc='sum')
```

```python
tornament_summary = home_tornament_summary.merge(away_tornament_summary, 
                                                 how='inner', left_index=True, right_index=True)
```

```python
tornament_summary.reset_index(inplace=True)
```

```python
tornament_summary = tornament_summary[tornament_summary['home_club_id'] == tornament_summary['away_club_id']]
```

```python
tornament_summary['total_points'] = tornament_summary['home_game_points'] + tornament_summary['away_game_points'] 
```

```python
tornament_summary.head(2)
```

```python
clubs = pd.read_csv('clubs.csv', usecols=['club_id','name','domestic_competition_id'])
```

```python
tornament_summary = tornament_summary.merge(clubs, how='left', left_on='home_club_id', right_on='club_id')
```

```python
tornament_summary.drop(['away_club_id'], axis=1, inplace=True)
```

```python
tornament_summary.rename(columns={'home_club_id':'club_id','name':'FC'}, inplace=True)
```

```python
league_identity = major_leagues.loc[:,['name','country_name','domestic_league_code']]
```

```python
tornament_summary = league_identity.merge(tornament_summary, how='left', left_on='domestic_league_code', 
                      right_on='domestic_competition_id')
```

```python
tornament_summary.drop(['domestic_league_code'], axis=1, inplace=True)
```

```python
tornament_summary = tornament_summary.sort_values(by=['country_name','season','total_points','goal_difference'], 
                                                  ascending=[True, True, False, False])
```

```python
tornament_summary = tornament_summary.loc[:,~tornament_summary.columns.duplicated()]
```

```python
tornament_summary.rename(columns={'name':'League'}, inplace=True)
tornament_summary.head(2)
```

```python
tornament_summary.to_csv('./modified/tornament_summary.csv', index=False)
```

```python
tornament_summary['group_rank'] = tornament_summary.groupby(
    ['season', 'League'])['total_points'].rank(ascending=False, method='first')
tornament_summary['group_rank'] = tornament_summary['group_rank'].astype('int')
```

```python
tornament_summary['goal_difference_y'] = tornament_summary['goal_difference_y'] * -1
```

```python
tornament_summary['goal_difference'] = tornament_summary['goal_difference_y'] + tornament_summary['goal_difference_x']
```

```python
tornament_summary.drop(['goal_difference_y','goal_difference_x'], axis=1, inplace=True)
```
